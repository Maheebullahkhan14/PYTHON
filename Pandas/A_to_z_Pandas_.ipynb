# Group by 
Allows you to group together rows based off of a column and perform an aggregate function on them

```python
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','James','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}

d1 = pd.DataFrame(data)
```

```python
d1
```

```python
bycomp = d1.groupby('Company')
```

```python
bycomp.mean()
```

```python
bycomp.sum()
```

```python
#std of sales per company
bycomp.std()
```

```python
#get the total sales of facebook company
bycomp.sum().loc['FB']
```

# Merging

```python
left = pd.DataFrame({'key':['K0','K1','K2','K3'],
                       'A':['A0','A1','A2','A3'],
                       'B':['B0','B1','B2','B3']})
```

```python
right = pd.DataFrame({'key':['K0','K1','K2','K3'],
                       'C':['C0','C1','C2','C3'],
                       'D':['D0','D1','D2','D3']})
```

```python
pd.merge(left,right, how = 'inner',on = 'key')
```

**Merging on multiple columns (keys)**

```python
left1 = pd.DataFrame({'key1':['K0','K1','K2','K3'],
                     'key2':['K0','K1','K0','K1'],
                       'A':['A0','A1','A2','A3'],
                       'B':['B0','B1','B2','B3']})
```

```python
right1 = pd.DataFrame({'key1':['K0','K1','K2','K3'],
                      'key2':['K0','K1','K0','K1'],
                       'C':['C0','C1','C2','C3'],
                       'D':['D0','D1','D2','D3']})
```

```python
# pd.merge(left,right, on = ['key1','key2'])
# Merge left and right
pd.merge(left1,right1, on = ['key1','key2'])
```
